```python
import pandas as pd
```

```python
df = pd.read_csv('ocorrencia.csv', sep=';', parse_dates=['ocorrencia_dia'], dayfirst=True)
df
```

```python
df.loc[df.ocorrencia_uf == '**', ['ocorrencia_uf']] = pd.NA
df
```

```python
df.replace(['**', '***', '****', '###!', '####'], pd.NA, inplace=True)
df
```

```python
df.isna().sum()
```

```python
df.fillna(1111, inplace=True)
df
```

```python
df.isna().sum()
```

```python
df['total_recomendacoes_bkp'] = df.total_recomendacoes
df.head()
```

```python
df.drop(['total_recomendacoes_bkp'], axis=1, inplace=True)
df.head()
```
